```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# loading dataset
data = pd.read_csv('Data\Training.csv')

# Selecting the top 10 most frequent symptoms excluding 'prognosis' and 'Unnamed: 133'
top_10_symptoms = data.drop(['prognosis', 'Unnamed: 133'], axis=1, errors='ignore').sum().sort_values(ascending=False).head(10)

# Customization dictionary
variables = {
    "size": (10, 8),
    "title": "Most Frequent Symptoms in the Dataset",
    "xlabel": "Frequency",
    "ylabel": "Symptoms",
    "ylabel_fontsize": 14,  # Adjust font size of y-axis labels
    "ylabel_fontweight": "bold"  # Make y-axis labels bold
}

# Plotting
plt.figure(figsize=variables['size'], facecolor='white')
barplot = sns.barplot(x=top_10_symptoms.values, y=top_10_symptoms.index, palette='viridis')
plt.title(variables['title'])
plt.xlabel(variables['xlabel'])
plt.ylabel(variables['ylabel'], fontsize=variables['ylabel_fontsize'], fontweight=variables['ylabel_fontweight'])
plt.tight_layout()

# Adjusting y-axis label properties individually
for item in barplot.get_yticklabels():
    item.set_fontsize(variables['ylabel_fontsize'])
    item.set_fontweight(variables['ylabel_fontweight'])

plt.show()
```
